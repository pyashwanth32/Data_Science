```python
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model=hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
    #return model(input)
    #return model(input).numpy()
    return model([tf.constant(input)]).numpy()
```

```python
forest=RandomForestClassifier(n_estimators=100, random_state=100)
forest.fit(xtrain,ytrain)
```

```python
pred = forest.predict(xvalid)
```

```python
pred.shape
```

```python
pred[:100]
#can't tune any more because its only 0 or 1, so it is bad for fire department
#    and the area under curve also is bad because if its percentage, then
#    it is better for fire department
```

```python
accuracy_forest=np.sum(pred==yvalid)/len(yvalid)*100
```

```python
accuracy_forest
```
